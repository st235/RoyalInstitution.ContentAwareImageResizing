```python
import cv2
import numpy as np
import pandas as pd
import matplotlib.pylab as plt
```

```python
def load_image(path):
  """Loads an image from hard drive using the given path.
  """
  return cv2.imread(path)
```

```python
def show_image(image):
  """Shows the image.
  """
  plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
```
